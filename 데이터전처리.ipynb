```python
import numpy as np
def remove_outliers_quantile(df, column, lower_quantile=0.05, upper_quantile=0.95):
    lower_bound = df[column].quantile(lower_quantile)
    upper_bound = df[column].quantile(upper_quantile)
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df

# '이용금액_업종기준' 열의 하위 5%와 상위 5%를 제거한 DataFrame 생성
filtered_df = remove_outliers_quantile(df, '이용금액_업종기준')

# 필터링된 데이터의 통계 정보 출력
print(filtered_df.describe())

# 히스토그램의 bin 개수 설정
bin_width = (filtered_df['이용금액_업종기준'].max() - filtered_df['이용금액_업종기준'].min()) / 100
bins = np.arange(filtered_df['이용금액_업종기준'].min(), filtered_df['이용금액_업종기준'].max() + bin_width, bin_width)

# 히스토그램 그리기
plt.figure(figsize=(12, 6))
plt.hist(filtered_df['이용금액_업종기준'], bins=bins, color='skyblue', edgecolor='black')
plt.title('Distribution of 이용금액_업종기준 (Filtered)')
plt.xlabel('이용금액_업종기준')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# '이용금액_업종기준' 열의 최소값과 최대값 확인
min_value = df['이용금액_업종기준'].min()
max_value = df['이용금액_업종기준'].max()
print(f"Min value: {min_value}, Max value: {max_value}")

# 히스토그램의 bin 개수 설정
bin_width = (max_value - min_value) / 100  # 예: 100개의 bin으로 나눔
bins = np.arange(min_value, max_value + bin_width, bin_width)

# 히스토그램 그리기
plt.figure(figsize=(12, 6))
plt.hist(df['이용금액_업종기준'], bins=bins, color='skyblue', edgecolor='black', log=True)
plt.title('Distribution of 이용금액_업종기준')
plt.xlabel('이용금액_업종기준')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
ratio = 0.80
cols = ['쇼핑_편의점_이용금액', '쇼핑_슈퍼마켓_이용금액', '쇼핑_마트_이용금액', '쇼핑_도소매_이용금액']

# 반복문을 사용하여 각 열의 값을 30%와 70%로 나누어 '이용금액_요식'과 '이용금액_쇼핑'에 반영
for col in cols:
    df['이용금액_요식'] += df[col] * ratio
    df['이용금액_쇼핑'] -= df[col] * ratio
    df['이용금액_쇼핑'] += df[col] * (1 - ratio)


# '쇼핑_편의점_이용금액'과 '쇼핑_슈퍼마켓_이용금액' 열 삭제
df = df.drop(columns=cols)

# '이용금액_업종기준'이 50만원 이상이고 150만원 미만인 행들만 필터링
filtered_df1 = df[(df['이용금액_업종기준'] >= 500000) & (df['이용금액_업종기준'] < 2000000)]

# 필터링된 데이터의 각 항목의 합계 및 비율 계산
total_usage = filtered_df1['이용금액_업종기준'].sum()
print(f"Total 이용금액_업종기준: {total_usage}")

# 각 항목의 총액 계산
shopping_total = filtered_df1['이용금액_쇼핑'].sum()
food_total = filtered_df1['이용금액_요식'].sum()
transport_total = filtered_df1['이용금액_교통'].sum()
medical_total = filtered_df1['이용금액_의료'].sum()
payment_total = filtered_df1['이용금액_납부'].sum()
education_total = filtered_df1['이용금액_교육'].sum()
leisure_total = filtered_df1['이용금액_여유생활'].sum()

# 각 항목의 비율 계산
shopping_ratio = (shopping_total / total_usage) * 100
food_ratio = (food_total / total_usage) * 100
transport_ratio = (transport_total / total_usage) * 100
medical_ratio = (medical_total / total_usage) * 100
payment_ratio = (payment_total / total_usage) * 100
education_ratio = (education_total / total_usage) * 100
leisure_ratio = (leisure_total / total_usage) * 100

# 결과 출력
print(f"Shopping Total: {shopping_total}, Ratio: {shopping_ratio:.2f}%")
print(f"Food Total: {food_total}, Ratio: {food_ratio:.2f}%")
print(f"Transport Total: {transport_total}, Ratio: {transport_ratio:.2f}%")
print(f"Medical Total: {medical_total}, Ratio: {medical_ratio:.2f}%")
print(f"Payment Total: {payment_total}, Ratio: {payment_ratio:.2f}%")
print(f"Education Total: {education_total}, Ratio: {education_ratio:.2f}%")
print(f"Leisure Total: {leisure_total}, Ratio: {leisure_ratio:.2f}%")


# 결과를 DataFrame으로 저장
results = {
    'Category': ['쇼핑', '요식', '교통', '의료', '납부', '교육', '여유생활'],
    'Ratio (%)': [shopping_ratio, food_ratio, transport_ratio, medical_ratio, payment_ratio, education_ratio, leisure_ratio]
}

results_df = pd.DataFrame(results)
sorted_results_df = results_df.sort_values(by='Ratio (%)', ascending=False)

# 결과 DataFrame 출력
print(sorted_results_df)
```
